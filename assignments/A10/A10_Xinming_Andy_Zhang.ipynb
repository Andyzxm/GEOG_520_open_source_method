```python
#graph 1 
USC= pd.read_csv("/content/drive/MyDrive/GEOG520_python_data/datasets/datasets/us_county_data.csv")
fig, axs = plt.subplots(1, 1)
sns.violinplot(ax=axs, x="SUB_REGION", y="per_internet", data=USC)
axs.grid(True)
axs.set_title("Internet Coverage by Sub Region", fontsize=20, color="#000000")
axs.set_xlabel("Sub-Region", fontsize= 15, color="#000000")
axs.set_ylabel("Internet coverage % ", fontsize=21, color="#000000")
axs.set_yticklabels(["20%", "30%","40%","50%", "60%", "70%", "80%", "90%", "100%"], fontsize=14, color="#000000")
axs.set_xticklabels(["ES Central", "Mountain", "WS Central", "Pacific", "New England", "South Atlantic", "EN Central", 
                     "WN Central", "Mid-Atlantic"], fontsize=10, color="#000000", rotation=45)
fig.patch.set_facecolor('#FFEBCD')
plt.show(fig)
```

```python
#graph 2 
fig2, axs =  plt.subplots(1, 1)
sns.scatterplot(ax=axs, x="med_income", y="per_internet", hue="SUB_REGION", size="med_income",sizes =(10, 50) ,data=USC)
axs.set_title("Distribution of median income and its relationship with internet coverage", fontsize=15, color="#000000")
plt.legend( fontsize='7', title_fontsize='10')
axs.grid(True)
axs.set_xlabel("Median Income $", fontsize= 15, color="#000000")
axs.set_ylabel("Internet coverage % ", fontsize=15, color="#000000")
axs.set_yticks([30,40,50,60,70,80,90,100])
axs.set_yticklabels(["30%","40%","50%", "60%", "70%", "80%", "90%","100%"], fontsize=14, color="#000000")
axs.set_xticks([20000, 40000, 60000,  80000, 100000,120000,140000])
axs.set_xticklabels(["$20,000", "$40,000", "$60,000", "$80,000", "$100,000", "$120,000","$140,000"], fontsize=10, color="#000000", rotation=45)
fig2.patch.set_color('#8B8378')
plt.show(fig2)
```

```python
#graph 3 
fig3, axs =  plt.subplots(1, 1)
sns.scatterplot(ax=axs, x="dem", y="tempmn", hue = "SUB_REGION", size="dem",sizes =(5,20), data = USC)
axs.set_title("Distribution of elevation and its relationship with temperature", fontsize=16, color="#000000")
plt.legend(fontsize='6', title_fontsize='8')
axs.grid(True)
axs.set_xlabel("Elevation", fontsize= 15, color="#000000")
axs.set_ylabel("Mean Temperature", fontsize=15, color="#000000")
axs.set_yticks([0,5,10,15,20,25])
axs.set_yticklabels(["0℃","5℃","10℃", "15℃", "20℃", "25℃"], fontsize=14, color="#000000")
axs.set_xticks([0, 500, 1000, 1500,2000,2500,3000,3500])
axs.set_xticklabels(["0m", "500m", "1000m", "1500m", "2000m","2500m","3000m","3500m"], fontsize=10, color="#000000", rotation=45)
fig3.patch.set_color('#00FFFF')
plt.show(fig3)
```
