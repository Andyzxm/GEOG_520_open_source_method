```python
from numpy.lib.function_base import append
#Task 1 : Create a scatterplot for all weapons to compare weight (“Weight”) and damage (“Damage”)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

skyrim=pd.read_csv("/content/drive/MyDrive/GEOG520_python_data/datasets/datasets/Skyrim_Weapons.csv")
eldenring=pd.read_csv("/content/drive/MyDrive/GEOG520_python_data/datasets/datasets/elden_ring_weapon.csv")
#filtering 5 weapon types from all weapons for skyrim
SFivetypes=skyrim.query('(Type == "Bow") | (Type == "Mace") | (Type == "Sword")| (Type == "War Axe") | (Type == "Warhammer") ')
SFivetypes
weapon_sca= plt.scatter(SFivetypes.Damage, SFivetypes.Weight)
plt.xlabel("Weapon Damage") 
plt.ylabel( "Weight")
plt.show(weapon_sca)
```

```python
#Task 2 Create a scatterplot for all weapons to compare cost (“Gold”) and damage(“Damage”)
weapon_sca2=plt.scatter(SFivetypes.Damage, SFivetypes.Gold)
plt.xlabel("Weapon Damage") 
plt.ylabel( "Gold")
plt.show(weapon_sca2)
```

```python
#Task 3 : Create a scatterplot for all weapons to compare weight (“Weight”) and damage (“Damage”). Differentiate the type of weapon (“Type”) using the point color
sns.scatterplot(x= "Weight", y= "Damage", hue= "Type", data= SFivetypes)
```

```python
#Task 4 : Create a grouped box plot to show the distribution of damage (“Damage”) grouped by weapon type (“Type”).
sns.boxplot(x= "Type", y = "Damage", data= SFivetypes)
```

```python
#Task 5 Create a grouped box plot to show the distribution of weapon weight (“Weight”) grouped by weapon type (“Type”).
sns.boxplot(x= "Type", y = "Weight", data= SFivetypes)
```

```python
#Task 6 Create a grouped violin plot to show the distribution of damage (“Damage”) grouped by weapon type (“Type”)
sns.violinplot(x= "Type", y= "Damage", data= SFivetypes)
```

```python
#Task 7 : Create a grouped violin plot plus grouped box plot to show the distribution of damage (“Damage”) grouped by weapon type (“Type”)
fig, axs =plt.subplots(2,1)
sns.violinplot(ax= axs[0],x= "Type", y= "Damage", data= SFivetypes)
sns.boxplot(ax=axs[1], x= "Type", y= "Damage", data= SFivetypes)
```

```python
#Task 8 Create a grouped violin plot plus grouped box plot to show the distribution ofdamage (“Damage”) grouped by weapon Category (“Category”)
fig, axs =plt.subplots(2,1)
sns.violinplot(ax= axs[0],x= "Category", y= "Damage", data= SFivetypes)
sns.boxplot(ax=axs[1], x= "Category", y= "Damage", data= SFivetypes)
```

```python
#Task 9 Create a scatterplot with weapon weight (“Weight”) mapped to the x-axisposition, damage (“Damage”) mapped to the y-axis position, type (“Type”) mapped to the point color, and weapon category (“Category”) mapped to the point symbol.
sns.scatterplot(x= "Weight", y = "Damage", hue="Type", style="Category", data= SFivetypes )
```

```python
#Task 10 Create a grouped box plot to show the distribution of physical damage (“Phy”) grouped by recoded weapon types (“Type”).
#reclassify the weapon types
eldenring['Type'] = eldenring['Type'].replace(['Colossal Sword', 'Curved Greatsword', 'Curved Sword', 'Heavy Thrusting Sword', 'Greatsword', 'Katana', 'Straight Sword'],'Sword')
eldenring['Type'] = eldenring['Type'].replace(['Axe', 'Greataxe'],'Axe')
eldenring['Type'] = eldenring['Type'].replace(['Hammer', 'Warhammer'],'Hammer')
E3types= eldenring.query('(Type == "Axe") | (Type == "Hammer") | (Type == "Sword") ')
box_plot, axs = plt.subplots(1,1)
sns.boxplot(ax = axs, x = "Type", y = "Phy", data = E3types)
axs.set_title("Distribution of physical damage based on weapon type in Eldenring", fontsize = 18)
axs.set_xlabel("Weapon type", fontsize = 15)
axs.set_ylabel("Physical damage", fontsize = 15)
plt.show(box_plot)
```

```python
#Task 11 Create a grouped box plot to show the distribution of weapon weight
#(“Wgt”) grouped by recoded weapon types (“Type”)
box_plot2 = sns.boxplot(x="Type", y="Wgt", data= E3types)
sns.boxplot(x="Type", y="Wgt", data= E3types)
box_plot2.set_title("distribution of weapon weight by weapon type")
box_plot2.set_xlabel("Weapon type")
box_plot2.set_ylabel("weight")
plt.show(box_plot2)
```

```python
#Task 12 : Create a scatterplot with weight (“Wgt”) mapped to the x-axis position,
#physical damage (“Phy”) mapped to the y-axis position, and recoded weapon type
#(“Type”) mapped to the point color. 
scat= sns.scatterplot(x= "Wgt", y = "Phy", hue="Type", data= E3types )
scat.set_title("distribution of weapon weight by weapon type")
scat.set_xlabel("Weapon type")
scat.set_ylabel("weight")
```

```python
#Task 13 :Create a histogram for just the swords (“Type”) showing the distribution of weapon weight (“Wgt”).
Eldenswrod= E3types.query('Type=="Sword"')
sns.histplot(x= "Wgt", data= Eldenswrod)
```

```python
#Task 14 : Create a kernel density plot for just the swords (“Type”) showing the distribution of weapon weight (“Wgt”)
Eldenswrod= E3types.query('Type=="Sword"')
sns.kdeplot(x= "Wgt", data= Eldenswrod)
```

```python
#Task 15 : Create a kernel density plot for all weapons grouped by type (“Type”) showing the distribution of weapon weight (“Wgt”). In other words, each weapon type should have its own density curve.
sns.kdeplot(x= "Wgt", hue= "Type", data= E3types)
```
