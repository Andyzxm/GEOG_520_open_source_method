```python
import numpy as np
import pandas as pd

#Task 1 What county has the highest recorded mean elevation?
USC=pd.read_csv("/content/drive/MyDrive/GEOG520_python_data/datasets/datasets/us_county_data.csv")
USC[USC.dem == USC.dem.max()]
```

```python
#Task 2 : What state has the lowest mean county elevation? 
USC[USC.dem== USC.dem.min()]
```

```python
#Task 3 : How many counties have a mean elevation greater than 1,000 meters?
USC.query('dem > 1000')
print(len(USC.query('dem > 1000')))
```

```python
#Task 4  How many counties have a mean elevation greater than 1,000 meters and a percent forest cover greater than 40%? 
USC.query('dem> 1000 and per_for>40')
print(len(USC.query('dem> 1000 and per_for>40')))
```

```python
#Task 5 : How many counties have a mean elevation lower than 100 meters and a percent developed larger than 35%?
USC.query('dem<100 and per_dev>35')
print(len(USC.query('dem<100 and per_dev>35')))
```

```python
#Task 6 : Which subregion of the country has the largest number of counties?
USC.groupby('SUB_REGION')["NAME"].count()
```

```python
#Task 7 which subregion of the country has the highest average county mean annual temperature?
USC.groupby('SUB_REGION')["tempmn"].max()
```

```python
#Task 8 Does Nebraska or Kansas have the highest average county percent cropland
NK= USC.query('STATE_NAME == "Nebraska" or STATE_NAME == "Kansas"')

NK.groupby("STATE_NAME")["per_crop"].mean()
```

```python
#Task 9 How many counties have a percent cropland greater than 25%, a mean temperature higher than 10Â°Celsius, and are not in the West South Central subregion?
USC.query('per_corp>25 and tempmn>10 and SUB_REGION!="W S Cen"')
print(len(USC.query('per_corp>25 and tempmn>10 and SUB_REGION!="W S Cen"')))
```

```python
#Task 10 Which county in the country has the highest percent cropland? 
USC[USC.per_crop == USC.per_crop.max()]
```

```python
#Task 11 Which county in the country has the highest percent developed land?
USC[USC.per_dev == USC.per_dev.max()]
```

```python
#Task 12 : Which subregion has the highest total population?

USC.groupby('SUB_REGION')["POPULATION"].sum()
```

```python
#Task 13 : Which state has a higher percent of households with internet access (all internet access, not just broadband), West Virginia or North Dakota?
WVvND = USC.query('STATE_NAME == "West Virginia" or STATE_NAME == "North Dakota"')
WVvND.groupby("STATE_NAME")["per_broadband" and "per_internet"].max()
```

```python
#Task 14:  Of all the counties with a mean elevation greater than 1,200 meters, which has the highest total annual precipitation?
USC1200= USC.query('dem > 1200')
USC1200[USC1200.precip == USC1200.precip.max()]
```

```python
# Task 15 What county has the largest difference between its annual lowest and highest temperatures?
USC["rad_dif"]=USC["tempmax"]-USC["tempmin"]
USC[USC.rad_dif == USC.rad_dif.max()  ]
```
